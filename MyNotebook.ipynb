```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import scipy.special as sp

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
df["Gender"].value_counts()
```

```python
df["Occupation"].value_counts()
```

```python
df.plot.scatter(x='Physical Activity Level', y='Sleep Duration')
```

```python
st.ttest_1samp(
    df["Sleep Duration"],
    7.5,
    alternative='greater'
).pvalue
```

```python
maleVar = np.var(df["Sleep Duration"].where(df["Gender"] == "Male"))
femaleVar = np.var(df["Sleep Duration"].where(df["Gender"] == "Female"))
print(maleVar, femaleVar)
```
