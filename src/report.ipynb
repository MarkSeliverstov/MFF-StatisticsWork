# Hypothesis Testing

## Hypothesis 1: Difference in Sleep Duration Between Genders

We are interested in investigating whether there is a significant difference in sleep duration between genders. To test this hypothesis, we will perform a hypothesis test, such as a t-test for independent samples, comparing the sleep duration of males and females in our dataset.

Null Hypothesis (H0): There is no significant difference in sleep duration between genders.\
Alternative Hypothesis (H1): There is a significant difference in sleep duration between genders.

We will conduct the appropriate statistical test and report the results, including the test statistic, p-value, and effect size (if applicable). If the p-value is less than a predetermined significance level (e.g., 0.05), we will reject the null hypothesis in favor of the alternative hypothesis.

```python
# Define the two groups based on gender
males = df[df['Gender'] == 'Male']['Sleep Duration']
females = df[df['Gender'] == 'Female']['Sleep Duration']

# Perform the t-test for independent samples
t_statistic, p_value = stats.ttest_ind(males, females)

# Set the significance level
alpha = 0.05

# Print the results
print("Hypothesis 1: Difference in Sleep Duration Between Genders")
print("Null Hypothesis (H0): There is no significant difference in sleep duration between genders.")
print("Alternative Hypothesis (H1): There is a significant difference in sleep duration between genders.")
print()
print("Test Statistic:", t_statistic)
print("p-value:", p_value)
print()

# Check if the null hypothesis should be rejected
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: There is a significant difference in sleep duration between genders.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference in sleep duration between genders.")

# Define the two groups based on gender
males = df[df['Gender'] == 'Male']['Sleep Duration']
females = df[df['Gender'] == 'Female']['Sleep Duration']

# Create a box plot
plt.boxplot([males, females], labels=['Male', 'Female'])
plt.title('Difference in Sleep Duration Between Genders')
plt.xlabel('Gender')
plt.ylabel('Sleep Duration')
plt.show()
```

## Hypothesis 2: Relationship Between Physical Activity Level and Sleep Quality

We want to explore the potential relationship between physical activity level and sleep quality. Specifically, we aim to determine whether individuals with higher physical activity levels tend to have better sleep quality scores. To test this hypothesis, we will use correlation analysis.

Null Hypothesis (H0): There is no significant correlation between physical activity level and sleep quality.\
Alternative Hypothesis (H1): There is a significant correlation between physical activity level and sleep quality.

We will calculate the correlation coefficient (e.g., Pearson's correlation coefficient) between physical activity level and sleep quality and assess its statistical significance. Additionally, we will provide a scatter plot to visualize the relationship between these variables.

The results of these hypothesis tests will help us draw conclusions about the differences in sleep duration between genders and the potential relationship between physical activity level and sleep quality in our dataset.

```python
# Calculate the correlation coefficient
correlation_coefficient, p_value = stats.pearsonr(df['Physical Activity Level'], df['Quality of Sleep'])

# Set the significance level
alpha = 0.05

# Print the results
print("Hypothesis 2: Relationship Between Physical Activity Level and Sleep Quality")
print("Null Hypothesis (H0): There is no significant correlation between physical activity level and sleep quality.")
print("Alternative Hypothesis (H1): There is a significant correlation between physical activity level and sleep quality.")
print()
print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
print()

# Check if the null hypothesis should be rejected
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: There is a significant correlation between physical activity level and sleep quality.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant correlation between physical activity level and sleep quality.")

# Create a scatter plot
plt.scatter(df['Physical Activity Level'], df['Quality of Sleep'])
plt.title('Relationship Between Physical Activity Level and Sleep Quality')
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.show()
```
